```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {'Year': [2020, 2021, 2022, 2023],
        'Sales': [200, 450, 700, 1000]}
df = pd.DataFrame(data)

print(df)

plt.plot(df['Year'], df['Sales'], marker='o')
plt.title("Sales Growth")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```
